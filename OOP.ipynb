
#introduction to OOP in python


OOP stands for Object-Oriented Programming. It is a programming paradigm that organizes code around objects, which are instances of classes. Think of a class as a blueprint and an object as a real-world instance of that blueprint.

For example, a Car class could have properties like color, make, and model and methods like start, stop, and accelerate. Each car object (like "my red Toyota Corolla") would be an instance of the Car class.

Key principles of OOP include:

`Encapsulation: Bundling data and the methods that operate on that data within a single unit (the class).`

`Abstraction: Hiding the complex implementation details and exposing only essential features of an object.`

`Inheritance: Creating new classes (subclasses) that inherit properties and methods from existing classes (parent classes).`

`Polymorphism: The ability of objects of different classes to respond to the same method call in their own way.`



#Benefits of OOP
Object-Oriented Programming (OOP) offers several benefits:

1. Modularity: Code is organized into reusable, self-contained objects (classes), making it easier to manage and maintain.

2. Flexibility: Objects can be easily modified and extended without affecting other parts of the code, promoting adaptability.

3. Maintainability: The structured nature of OOP code simplifies maintenance and debugging.

4. Reusability: Classes can be reused in different parts of the project or in other projects, saving development time.

#4- Encapsulation and Abstraction

Encapsulation and Abstraction are key principles in Object-Oriented Programming (OOP).

##Encapsulation:

1.  Bundling: Think of it as wrapping data (attributes) and the methods that work on that data within a single unit, which is the class.
2. Protection: Encapsulation controls access to the internal state of an object. You can restrict direct modification of attributes from outside the class.  
3.Implementation Hiding: It hides the internal details of how an object works, exposing only necessary interfaces.





Example in your code:



    #class car:
    ... (other code) ...
        def drive(self):
            print(f"you started driving the {self.model}")   
        def stop(self):
            print(f"you stopped driving the {self.model}")
    # ... (other code) ...

Here, the drive and stop methods encapsulate the logic of how a car starts and stops. The user of the car class doesn't need to worry about the internal implementation, they just call these methods.

##Abstraction:

Simplification: It focuses on presenting essential features of an object while hiding the complex underlying details.
Interface: Abstraction defines a clear contract for how an object should be used, without revealing how it fulfills that contract.
Example (not explicitly in your code, but how you could apply it):

Imagine you want to add a feature to calculate the fuel efficiency of a car. You could create an abstract method calculate_fuel_efficiency in the car class. Each subclass of car (like ElectricCar, PetrolCar) would then implement this method in their own specific way. The user interacting with these objects only needs to know about the calculate_fuel_efficiency method, not the detailed calculations behind it.

`Key Differences:`

`Encapsulation is about bundling data and methods together and controlling access.
Abstraction is about simplifying the interaction with an object and hiding complexity.`

#5- Inheritance

###Inheritance is one of the core concepts of Object-Oriented Programming (OOP). It allows you to create a hierarchy of classes where a new class (subclass or derived class) can inherit properties and behaviors (methods) from an existing class (parent class or base class). This promotes code reusability and a well-organized class structure.

````python
#inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement abstract method")
        #or pass  # Placeholder for the method as it is will be used diff for each subclass
class Dog(Animal):#means that the class Dog is an inherant of class Animal attributes and methods/functions
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):#means that the class Cat is an inherant of class Animal attributes and methods/functions
    def speak(self):
        return f"{self.name} says Meow!"

# Create instances of Dog and Cat  object creation
dog = Dog("husk")#Dog class has argument name inherited from class Animal
cat = Cat("Tom")#Cat class has argument name inherited from class Animal

# Call the speak method
print(dog.speak())  # Output: Buddy says Woof!
print(cat.speak())  # Output: Whiskers says Meow!
````

#6-Composition

###Composition is a way to build complex objects by combining simpler objects. Think of it like building with LEGOs â€“ you take individual bricks (smaller objects) and put them together to create a larger structure (a complex object).

###In OOP terms, a class (the complex object) contains instances of other classes (the simpler objects) as attributes. This creates a "has-a" relationship. For instance, a Car has an Engine and has Wheels.

````python
class Engine:
    def __init__(self, horsepower):#the Engine class has horsepower attribute
        self.horsepower = horsepower

    def start(self):
        return f"Engine started with horsepower: {self.horsepower}"

class Wheel:
    def __init__(self, wheel_size):
        self.wheel_size = wheel_size

    def rotate(self):
        return  f"Wheel with size {self.wheel_size}  is rotating"

class Car:#composition simply implies that the car class has the attributes and methods of the previous classes and their functualaties
    def __init__(self, brand, horsepower, wheel_size):#this just acts a placeholders for the attributes of any object(car) created later
        self.brand = brand
        self.engine = Engine(horsepower)
        # Car has an Engine(class) of attribute(horsepower),the assigned value is the horsepower attribute which is an  arg  in def__init__ of Engine class
        self.wheels = [Wheel(wheel_size) for wheel in range(4)]  # Car has 4 Wheels

    def start_engine(self):
        return self.engine.start()

    def drive(self):
        return [wheel.rotate() for wheel in self.wheels]#to create a for loop accounting that the car has 4 wheels

# Creating a car object
my_car = Car("Toyota", 200, 16)

# Starting the engine
print(my_car.start_engine())  # Output: Engine started with horsepower: 200

# Driving the car
print(my_car.drive())  # Output: ['Wheel with size 16 is rotating', 'Wheel with size 16 is rotating', 'Wheel with size 16 is rotating', 'Wheel with size 16 is rotating']
````

#7- Polymorphism
Polymorphism in OOP allows objects of different classes to respond to the same method call in their own way.

````python
class animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass  # Placeholder for the method as it is will be used diff for each subclass
        #or raise NotImplementedError("Subclass must implement abstract method")

class dog(animal):#means that the class Dog is an inherant of class Animal attributes and methods/functions
    def speak(self):
        return  f" {self.name} says Woof!"

class cat(animal):
    def speak(self):
        return  f" {self.name} says Meow!"

animals = [dog("husk"), cat("Tom")]#they must be filled in with name arg(attribute) that is inherited from class animal
for i in animals:
    print(i.speak())  # Output will be "Woof!" and then "Meow!"
````

In this example, both Dog and Cat classes inherit from Animal and override the speak method. When you call speak on each animal, it executes the version specific to that class, demonstrating polymorphism.

This lets you write code that interacts with objects in a generic way, **without needing to know their specific class beforehand**.

#8- Special (Magic) Methods

Special methods (also called magic methods or dunder methods) in Python are predefined methods with names surrounded by double underscores (e.g., __init__, __str__, __add__). They provide a way to customize how your objects interact with built-in operators and functions.

###Common Magic Methods

Here are a few examples of common magic methods and their uses:

__init__(self, ...): Constructor method, called when an object is created.âœ…

__str__(self): Returns a human-readable string representation of an object.

__repr__(self): Returns a more developer-friendly string representation, often used for debugging.

__eq__(self, other): Defines how to compare two objects for equality.

__len__(self): Returns the length of an object (if applicable).

__add__(self, other): Defines how to add two objects together.
